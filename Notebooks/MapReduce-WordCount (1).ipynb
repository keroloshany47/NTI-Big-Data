```python
from functools import reduce
import string
```

```python
translator = str.maketrans('', '', string.punctuation)
```

```python
'test!*'.translate(translator)
```

```python
words_lst = open('shakespeare.txt')\
                .read()\
                .split()
```

```python
words_lst
```

```python
unified_lst = list(map(lambda word: word.translate(translator)\
                                       .lower()\
                                       , words_lst))


#def cleanse(word):
#    word = word.translate(translator)
#    word = word.lower()
#    return word
```

```python
unified_lst
```

```python
map_lst = list(map(lambda word: dict([[word, 1]]), unified_lst))
```

```python
map_lst
```

```python
def reducer(i, j):
    for k in j: i[k] = i.get(k, 0) + j.get(k, 0)
    return i
```

```python
reduce(reducer,map_lst)
```
